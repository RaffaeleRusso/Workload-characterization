```python
import threading
import time
import os 
import pandas as pd
from datetime import datetime
import subprocess
```

```python
commands = [
    "grep VmPeak /proc/"+str(apache_pid)+"/status | awk ' { print $2}' ",
    "grep VmRSS /proc/"+str(apache_pid)+"/status | awk ' { print $2}' ",
    "grep VmSize /proc/"+str(apache_pid)+"/status | awk ' { print $2}' ",
    "grep VmHWM/ proc/"+str(apache_pid)+"/status | awk ' { print $2}' ",
    "grep VmPTE /proc/"+str(apache_pid)+"/status | awk ' { print $2}' ",
    "grep Threads /proc/"+str(apache_pid)+"/status | awk ' { print $2}' ",
    "grep MemFree /proc/meminfo | awk ' { print $2}'",
    "grep Buffers /proc/meminfo | awk ' { print $2}'",
    "grep Cached /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep Active /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep Inactive /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep Dirty /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep Writeback /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep AnonPages /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep Mapped /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep Slab /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep PageTables /proc/meminfo |  head -1 | awk ' { print $2}'",
    "grep Committed_AS /proc/meminfo |  head -1 | awk ' { print $2}'",
    "head -1 /proc/sys/fs/file-nr | awk ' {print $1}' ",
    "cat /proc/sys/fs/file-nr | awk ' { print $1 } '"
]
```

```python
outputs = [[] for _ in range(len(commands)+1)] #the last one is for the timestamp
```

```python
def fun(i):
    val = subprocess.check_output(commands[i],cwd = "/",shell=True).decode('utf-8').replace("\n",'')
    outputs[i].append(val)
    
```

```python
n_it = 100
for it in range(n_it):
    threads = [threading.Thread(target = fun,args = (i,))for i in range(len(commands))]
    outputs[len(commands)].append(str(datetime.now())[-15:-7])
    for thread in threads:
        thread.start()
        print(f'Iterazione: {it} Thread_name = {thread.name}')
    thread.join()
    time.sleep(1)
```

```python
df = pd.DataFrame(outputs)
df = df.transpose()
df.columns=headers
```

```python
if not os.path.isfile('server.csv'):
   df.to_csv('server.csv',index=False)
else: 
   df.to_csv('server.csv', index = False,mode='a', header=False)
```
